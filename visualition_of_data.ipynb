```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tourists_new = pd.read_excel('clean_tourist_data (7).xlsx')
```

```python
plt.figure(figsize=(10,7))
sns.set_style("white")
sns.barplot(data=tourists_new, x = "Number_Of_Children", y= "Spending_Money($)", estimator=np.mean, palette= "flare")
plt.title('Average Money Spent by Number of Children', fontweight="bold",fontsize=20)
plt.xlabel("Number of Children", fontsize= 12)
plt.ylabel("Average Money Spent ($)", fontsize= 12)
sns.despine()
```

```python
plt.figure(figsize=(10,7))
sns.set_style("white")
sns.barplot(data=tourists_new, x = "Country", y= "Number_Of_Children", estimator=np.mean, palette= "crest")
plt.title('Average Number of Children by Country', fontweight="bold",fontsize=20)
plt.xlabel("")
plt.xticks(fontsize= 12)
plt.ylabel("Average Number of Children", fontsize= 12)
sns.despine()

```

```python
plt.figure(figsize=(10,7))
sns.set_style("white")
sns.barplot(data=tourists_new, x = "Education_Level", y= "Spending_Money($)", estimator=np.mean, palette= "rocket_r")
plt.title('Average Money Spent by Education Level', fontweight="bold",fontsize=20)
plt.xlabel("Education Level")
plt.ylabel("Average Money Spent ($)", fontsize= 12)
sns.despine()
```

```python
plt.figure(figsize=(10,7))
sns.set_style("white")
sns.scatterplot(data=tourists_new, x = "Monthly_Income($)", y= "Spending_Money($)", color= "gray")
plt.title('Money Spent by Monthly Income', fontweight="bold",fontsize=20)
plt.xlabel("Monthly Income ($)", fontsize= 12)
plt.ylabel("Money Spent ($)", fontsize= 12)
sns.despine()
```

```python
plt.figure(figsize=(10,7))
sns.set_style("white")
sns.scatterplot(data=tourists_new, x = "Weight(kg)", y= "Spending_Money($)", color= "orange")
plt.title('Money Spent by Weight', fontweight="bold",fontsize=20)
plt.xlabel("Weight (kg)", fontsize= 12)
plt.ylabel("Money Spent ($)", fontsize= 12)
sns.despine()
```

```python
plt.figure(figsize=(10,7))
sns.set_style("white")
custom_palette = ["#FF0000", "#00FFFF", "#FFFF00"]
sns.barplot(data=tourists_new, x = "Spending_Money($)", y= "Country", estimator=np.mean, palette= custom_palette)
plt.title('Average Money Spent by Country', fontweight="bold",fontsize=20)
plt.xlabel("Average Money Spent ($)", fontsize= 12)
plt.ylabel("", fontsize= 12)
plt.yticks(fontsize= 12)
sns.despine()
```

```python
plt.figure(figsize=(7,10))
sns.set_style("white")
sns.boxplot(data= tourists_new, y= "Spending_Money($)")
plt.title('Money Spent', fontweight="bold",fontsize=20)
plt.ylabel("Money Spent ($)", fontsize= 12)
sns.despine()
```
